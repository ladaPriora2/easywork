В компетенциях 7-10 приведены специфичные группы умений, которые не свойственны для других компетенций. Принимается решение удалить столбцы, связанные с этими группами компетенций.

```python
data_all[~data_all['компетенция 7'].isnull()]['компетенция 7'].unique()
```

Анкета, приведённая в примере, $-$ это образец тех данных, которые будет принимать на вход итоговая система. Нужно обратить внимание, что в анкете представлены группы навыков и показатель их важности в процентном соотношении.

Необходимо выделить такие группы в наших данных.

```python
data_all['название'].unique()
```

```python
# необходимо стандартизировать названия групп навыков и умений
def text_preprocessing(text):
    import re
    import string
    text = str(text)
    text = re.sub(r'\n|\r|\b', ' ', text, flags=re.S)
    text = ''.join([c for c in text if c not in string.punctuation]).lower().strip()
    text = ''.join([c for c in text if c not in string.digits])
    text = re.sub(r" +", " ", text, flags=re.S)
    return text

for i in range(1, 11):
    data_all[f'компетенция {i}'] = data_all[f'компетенция {i}'].apply(text_preprocessing)
data_all.head(5)
```
